# Comparison of clustering algorithms

- kmeans
- hierarchical clustering


```python
import pandas as pd
import df_manipulation_functions as dfmf
import math
import matplotlib.pyplot as plt
import numpy as np 
```

```python
path_mds = 'Flow_mds/additional_features_mds_flow.pkl'
path_labels_kmeans = 'Labels\Kmeans\cluster_labels_kmeans_additional_flow.csv'
path_labels_hierarchical ='Labels\Hierarchical\cluster_labels_hierachical_additional_flow.csv'

#import mds representation of feature both
mds_rep = pd.read_pickle(path_mds)

cluster_labels_kmeans = pd.read_csv(path_labels_kmeans)
cluster_labels_hierarchical = pd.read_csv(path_labels_hierarchical)

```

```python

clusters = np.arange(2,10)

for cluster_num in clusters :
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.suptitle(str(cluster_num) + ' clusters')
    ax1.scatter(mds_rep['x_MDS'], mds_rep['y_MDS'], c=cluster_labels_hierarchical[str(cluster_num) + ' clusters'])
    ax2.scatter(mds_rep['x_MDS'], mds_rep['y_MDS'], c=cluster_labels_kmeans[str(cluster_num) + ' clusters'])
    ax1.set_title('hierarchical')
    ax2.set_title('kmeans')



```
