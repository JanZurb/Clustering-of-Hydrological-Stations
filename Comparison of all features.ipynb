```python
import pandas as pd
import df_manipulation_functions as dfmf
import matplotlib.pyplot as plt
import numpy as np
```

```python
all_features_flow = pd.read_pickle('Flow_features/all_features_flow.pkl')
all_features_temp = pd.read_pickle('Temp_features/all_features_temp.pkl')

all_features_both = pd.merge(all_features_flow, all_features_temp,  on=['Stationsnummer'])
```

```python
#plot all features individually on a graph from df all_features_both

#create array with nums form 1 to 53
x = np.arange(1,54)
zeros = np.zeros(75)

for column in all_features_both:
    plt.figure(figsize=(10,1))
    plt.scatter(all_features_both[column],zeros, label=column)
    plt.legend()
    plt.show()


```

```python
all_features_both_normalized = dfmf.normalize_data(all_features_both)

x = np.arange(1,54)
zeros = np.zeros(75)

for column in all_features_both_normalized:
    plt.figure(figsize=(10,1))
    plt.scatter(all_features_both_normalized[column],zeros, label=column)
    plt.legend()
    plt.show()
    
```
